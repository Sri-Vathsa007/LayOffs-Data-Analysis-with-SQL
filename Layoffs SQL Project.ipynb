# Basic EDA

### Retrieving all the columns

```python
%%sql

SELECT * FROM layoff_data;
```

### Retrieving Distinct Companies

```python
%%sql

SELECT DISTINCT(Company) FROM layoff_data;
```

### Grouping by the companies

```python
%%sql

SELECT Company, COUNT(Company) AS 'Count of companies in this dataset' FROM layoff_data
GROUP BY Company
ORDER BY 2 DESC;
```

### Grouping by each country and seeing its laidoff count

```python
%%sql

SELECT Country, SUM(Laid_off_Count) AS '# Laid Off' FROM layoff_data
GROUP BY Country
ORDER BY 2 DESC;
```

### Top 3 tech gaints who has laid off most employees

```python
%%sql

SELECT Company, SUM(Laid_Off_Count) AS '# of Laid Off' FROM layoff_data
GROUP BY Company
ORDER BY 2 DESC LIMIT 3 
```

### Top 3 tech gaints who has laid off most employees with country

```python
%%sql

SELECT Company, Country ,SUM(Laid_Off_Count) AS '# of Laid Off' FROM layoff_data
GROUP BY Company,Country
ORDER BY 3 DESC LIMIT 3 
```

### Top 3 Industries who has laid off most employees with country

```python
%%sql

SELECT Industry, Country ,SUM(Laid_Off_Count) AS '# of Laid Off' FROM layoff_data
GROUP BY Company,Country
ORDER BY 3 DESC LIMIT 3 
```

# Advanced EDA

### Laid off Info by year by each company and country

```python
%%sql

SELECT strftime('%Y', Date), Company, Country, SUM(Laid_Off_Count) AS "# of Laid off" FROM layoff_data
GROUP BY strftime('%Y', Date), Company
ORDER BY 1,4 DESC;
```

### Ranking the Company that has made more lay off by the year

```python
%%sql

SELECT strftime('%Y', Date) as Year, Company, SUM(Laid_Off_Count) AS "# of Laid off", RANK() OVER(PARTITION BY strftime('%Y', Date) ORDER BY Date DESC) AS Rank
FROM layoff_data
GROUP BY strftime('%Y', Date)
```

### Ranking the countries that has done the most layoffs

```python
%%sql

SELECT Country, SUM(Laid_Off_Count), DENSE_RANK() OVER(PARTITION BY Country ORDER BY Laid_Off_Count DESC) AS Rank
FROM layoff_data
GROUP BY Country
ORDER BY 2 DESC
```
